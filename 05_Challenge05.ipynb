Problem<br>
This is a design problem. Your next project supports an accounting application and you’ve being assigned to implement a class which objective is to provide the implementation of a class with 1/10000 precision.<br>
Constraints<br>
1. Both, the inputs and output to your class are strings. Your constructor only admits strings representing numbers (int or float), including negative num- bers.
2. Other inputs are acceptable, but not in the constructor. This means you have to cretate your object from floats also.
3. You must support arithmetical operations with other object than your own type. Your type must interact with floats and ints as well.
4. Choose your internal variables wisely, however you’re not allowed to use Dec- imal or float objects.
5. Follow better programming and documentation practices with your submis- sion.
6. Provide effective test driven practices with your code.<br>
Bonus:<br>
Implement a derived class that allows converting your accounting amount into en- glish words, i. e, $124,980.76 translates into "One hundred and twenty four thou- sand nine hundred and eighty 76/100"
Write a document explaining your design decisions on why you chose the in- ternal variables the way you presented your implementation.

```python
import math

class Account:

    pres = 4  # Class precision

    # Initialize string input (must represent a number)
    def __init__(self, text: str):
#        self.number = self.floattoint(float(text))
        self.number = float(text)

    # Create object from numeric input
    @classmethod
    def numinp(cls, number: float):
        return cls(str(number))

    # DATA CONVERSION

    # Internal numbers
    def floattoint(self,number):
        return round(number*10**self.pres,0)
    def inttofloat(self,number):
        return round(number/(10**self.pres),self.pres)
    
    # Class-precision float
    def __float__(self):
        return round(self.number, self.pres)
    def __int__(self):
        return math.trunc(self.number)

    # Class-precision output
    def myout(self,number):
        return f"{number:.{self.pres}f}"

    # ARITHMETIC OPERATIONS

    # Addition +
    def __add__(self, other):
        x = self.number + float(other)
        return self.myout(x)

    def __radd__(self, other):
        return self.__add__(float(other))

    # Subtraction -
    def __sub__(self, other):
        x = self.number - float(other)
        return self.myout(x)

    def __rsub__(self, other):
        x = float(other) - self.number
        return self.myout(x)

    # Multiplication *
    def __mul__(self, other):
        x = self.number * float(other)
        return self.myout(x)

    def __rmul__(self, other):
        return self.__mul__(float(other))

    # Floor division //
    def __floordiv__(self, other):
        x = self.number // float(other)
        return self.myout(x)

    def __rfloordiv__(self, other):
        x = float(other) // self.number
        return self.myout(x)

    # True division /
    def __truediv__(self, other):
        x = self.number / float(other)
        return self.myout(x)

    def __rtruediv__(self, other):
        x = float(other) / self.number
        return self.myout(x)

    # Modulo %
    def __mod__(self, other):
        x = self.number % float(other)
        return self.myout(x)

    def __rmod__(self, other):
        x = float(other) % self.number
        return self.myout(x)

    # Power **
    def __pow__(self, other):
        x = self.number ** float(other)
        return self.myout(x)

    def __rpow__(self, other):
        x = float(other) ** self.number
        return self.myout(x)
```
